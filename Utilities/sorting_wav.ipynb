```python
import shutil,os
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
csvpath = r'H:\Respiratory_Sound_Database\csv\tachdata\all_data_Tachypnea.csv'
wav_files = pd.read_csv(csvpath,header=None, names=['files','sound'])
wav_files
```

```python
wav_files['sound'].unique()
```

```python
dfs = wav_files.copy()
for x in range(0,len(dfs)):
    file = dfs['files'][x]
    file = file[0:-4]+'.png'
    spath = Path('H:/Respiratory_Sound_Database/frequency_spectrograms_ffmpeg/')
    path_start = spath/file
    path_end = spath/dfs['sound'][x]
    shutil.move(str(path_start),str(path_end))
    print('Moving From: '+ str(path_start))
    print('To: '+ str(path_end))
    print()
```
