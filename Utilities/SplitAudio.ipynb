```python
inputDirectory = r'C:\Users\Chinmayi\Anaconda\Datasets\LungSounds_db\Respiratory_Sound_Database\audio_and_txt_files_input'
outputDirectory = r'C:\Users\Chinmayi\Anaconda\Datasets\LungSounds_db\Respiratory_Sound_Database\audio_and_txt_files_output'
summaryFilePath = r'C:\Users\Chinmayi\Anaconda\Datasets\LungSounds_db\Respiratory_Sound_Database\audio_and_txt_files_output\summary.txt'
```

```python
from pydub import AudioSegment
import os
import pandas as pd
```

```python
def splitAndDump(audioFile, metaDataFile):
    print('working on audio: ' + audioFile + ' and metafile: ' + metaDataFile)
    recording_annotations = pd.read_csv(metaDataFile,
                                        names=['Start', 'End', 'Crackles', 'Wheezes'], delimiter='\t')

    recording_audio = AudioSegment.from_wav(audioFile)

    for i in range(len(recording_annotations.index)):
        row = recording_annotations.loc[i]
        start = row['Start']
        end = row['End']
        crackles = int(row['Crackles'])
        wheezes = int(row['Wheezes'])
        audio_chunk = recording_audio[int(start*1000) : int(end*1000)]
        audio_chunk_filename = os.path.basename(audioFile).replace(".wav", str(i) +".wav")
        audio_chunk.export(os.path.join(outputDirectory, audio_chunk_filename), format="wav")
        summaryFile = open(summaryFilePath, "a")
        summaryFile.write(audio_chunk_filename + '\t'+ str(crackles) + '\t' + str(wheezes) + "\n")
        summaryFile.close()
```

```python
for filename in os.listdir(inputDirectory):
    if filename.endswith(".wav"):
        audioFile = os.path.join(inputDirectory, filename)
        metaDataFile = os.path.join(inputDirectory, filename.replace(".wav",".txt"))
        splitAndDump(audioFile, metaDataFile)
    else:
        continue
```
