## Converting audio files to spectrograms

```python
import os
```

```python
os.chdir(r'H:\Respiratory_Sound_Database\audio_and_txt_files_output')
for filename in os.listdir(r'H:\Respiratory_Sound_Database\audio_and_txt_files_output'):
    if (filename.endswith(".wav")):
        print(filename)
        os.system('ffmpeg -i {i}.wav -lavfi showspectrumpic=s=128x64:legend=false spectrograms\{i}.png'.format(i = filename[0:-4]))
    else:
        continue
```
