```python
inputDirectory = '../cracklesinputsplit/Respiratory_Sound_Database/Respiratory_Sound_Database/audio_and_txt_files_input'
outputDirectory = '../cracklesinputsplit/Respiratory_Sound_Database/Respiratory_Sound_Database/audio_and_txt_files_output'
summaryFilePath = '../cracklesinputsplit/Respiratory_Sound_Database/Respiratory_Sound_Database/audio_and_txt_files_output/summary.txt'
```

```python
from pydub import AudioSegment
import os
import pandas as pd
import numpy as np
import librosa
```

```python
def stretch(data, rate=np.random.uniform(low=0.5,high=1.7)):
    data = librosa.effects.time_stretch(data, rate)
    return data
```

```python
def splitAndDump(audioFile, metaDataFile):
    print('working on audio: ' + audioFile + ' and metafile: ' + metaDataFile)
    recording_annotations = pd.read_csv(metaDataFile,
                                        names=['Start', 'End', 'Crackles', 'Wheezes'], delimiter='\t')

    recording_audio = AudioSegment.from_wav(audioFile)

    for i in range(len(recording_annotations.index)):
        row = recording_annotations.loc[i]
        start = row['Start']
        end = row['End']
        crackles = int(row['Crackles'])
        wheezes = int(row['Wheezes'])
        audio_chunk = recording_audio[int(start*1000) : int(end*1000)]
        audio_chunk_filename = os.path.basename(audioFile).replace(".wav", str(i) +".wav")
        if ((crackles == 1) & (wheezes == 0)):
            sortedOutputDirectory = outputDirectory + "/coarse"    
        elif (crackles == 0) & (wheezes == 1):
            sortedOutputDirectory = outputDirectory + "/fine"
        elif (crackles == 0) & (wheezes == 0):
            sortedOutputDirectory = outputDirectory + "/healthy"
        else :
            sortedOutputDirectory = outputDirectory + "/both" #Not enough files, so not considering
        print ("length of audio: " +str(len(audio_chunk)))
        pad_ms = 5000

        while (len(audio_chunk) <= pad_ms) :
            audio_chunk = audio_chunk + audio_chunk
            print ("length of audio: " +str(len(audio_chunk)))

        if len(audio_chunk)>=pad_ms:
            audio_chunk = audio_chunk[:pad_ms]
            print ("length of audio: " +str(len(audio_chunk)))

        audio_chunk.export(os.path.join(sortedOutputDirectory   , audio_chunk_filename), format="wav")
        summaryFile = open(summaryFilePath, "a")
        summaryFile.write(audio_chunk_filename + '\t'+ str(crackles) + '\t' + str(wheezes) + "\n")

        summaryFile.close()
```

```python
for filename in os.listdir(inputDirectory):
    if filename.endswith(".wav"):
        audioFile = os.path.join(inputDirectory, filename)
        metaDataFile = os.path.join(inputDirectory, filename.replace(".wav",".txt"))
        splitAndDump(audioFile, metaDataFile)
    else:
        continue
```
